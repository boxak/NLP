```r
PosTable <- RcppMeCab::pos(str,format = "data.frame") ### str을 형태소 분석해서 PosTable이라는 임시 테이블에 저장
PosTable
```

```r
tempVector <- vector() ###형태소 원형을 담을 벡터
size <- nrow(PosTable)
result <- '' ### 벡터에 담긴 원소들을 콤마로 구분해서 result라는 문자열에 넣을 것임.
source(file = "findStem.R",encoding = "UTF-8") ### 네이버 사전에서 데이터를 끌어올 함수
```

```r
for(i in 1:size){
    ###품사가 일반명사, 고유명사이면 그대로 tempVector에 담는다.
    if(PosTable[i,5]=="NNG" | PosTable[i,5]=="NNP"){
      tempVector <- c(tempVector,as.character(PosTable[i,4]))
    }
    ###품사가 동사원형이면 -> 즉, "다"만 뒤에 붙이면 되는, 어간이 보존되어 있는 경우 "다"만 붙인다.
    else if(PosTable[i,5]=="VV"){
      addStr <- paste0(as.character(PosTable[i,4]),"다")
      tempVector <- c(tempVector,addStr)
    }
    ###어간의 원형이 변형되어 있으면, 즉 POS = VV+ 꼴이면 네이버 사전을 참조한다.
    else if(str_detect(PosTable[i,5],"VV")){
      text <- findStem(as.character(PosTable[i,4]))
      tempVector <- c(tempVector,text)
    }
}
tempVector
```

```r
### tempVector 에 아무것도 담기지 않으면 인덱스 초과 에러가 나므로 if문으로 길이가 0인지 판단.
### result 문자열에 콤마를 붙여서 예쁘게 넣는다.
if(length(tempVector)>0){
    result <- tempVector[[1]]
    if(length(tempVector)>1){
      for(i in 2:length(tempVector)){
        result <- paste(result,tempVector[[i]],sep = ", ")
      }
    }
}
result
```
